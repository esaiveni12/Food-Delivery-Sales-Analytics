```python
import pandas as pd

df = pd.read_csv('food_delivery.csv')

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Revenue'] = df['Quantity'] * df['Price']

df.head()
```

```python
top_items = df.groupby('Food_Item')['Quantity'].sum()

print(top_items)
```

```python
import matplotlib.pyplot as plt

top_items.plot(kind='bar')

plt.title('Top Selling Food Items')
plt.xlabel('Food Item')
plt.ylabel('Quantity Sold')

plt.show()
```

```python
df['Location'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Location-wise Orders')
plt.ylabel('')
plt.show()
```

```python
revenue = df.groupby('Food_Item')['Revenue'].sum()

print("----- BUSINESS INSIGHTS -----")
print("Most Popular Food Item:", top_items.idxmax())
print("Highest Revenue Food Item:", revenue.idxmax())
print("Total Revenue:", df['Revenue'].sum())
```
